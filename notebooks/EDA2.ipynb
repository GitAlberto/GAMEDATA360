```python
def parse_genre_final(x):
    """Version FINALE et ROBUSTE - Gère TOUS les cas."""
    
    # 1) Valeurs nulles ou vides
    if x is None or (isinstance(x, float) and pd.isna(x)):
        return []
    
    # 2) Déjà une liste Python propre
    if isinstance(x, list):
        # Nettoyer chaque élément de la liste
        cleaned = []
        for v in x:
            if v:
                v_clean = str(v).strip().lower()
                # Enlever les caractères spéciaux résiduels
                v_clean = re.sub(r"^[\['\"\]]+|[\['\"\]]+$", "", v_clean)
                if v_clean and v_clean not in ['aucun', 'none', 'nan']:
                    cleaned.append(v_clean)
        return cleaned
    
    # 3) Conversion en string
    x = str(x).strip().lower()
    
    # 4) Chaîne vide ou invalide
    if not x or x in ['[]', '', 'nan', 'none', 'aucun']:
        return []
    
    # 5) Essayer d'évaluer comme liste Python d'abord
    try:
        # Remplacer les quotes simples par doubles pour ast.literal_eval
        x_eval = x.replace("'", '"')
        parsed = ast.literal_eval(x_eval)
        if isinstance(parsed, list):
            cleaned = []
            for v in parsed:
                v_clean = str(v).strip().lower()
                v_clean = re.sub(r"^[\['\"\]]+|[\['\"\]]+$", "", v_clean)
                if v_clean and v_clean not in ['aucun', 'none', 'nan']:
                    cleaned.append(v_clean)
            return cleaned
    except:
        pass
    
    # 6) NETTOYAGE BRUTAL - Enlever TOUS les caractères spéciaux
    x = re.sub(r"[\[\]'\"]", "", x)  # Supprimer [ ] ' "
    
    # 7) Séparer par virgules
    genres = [g.strip() for g in x.split(',') if g.strip()]
    
    # 8) Filtrer les valeurs invalides
    invalid = {'aucun', 'none', 'nan', ''}
    genres = [g for g in genres if g and g not in invalid]
    
    return genres
```

```python
# Nettyage de la colonne Estimated owners
def parse_estimated_sales(x):
    """Convertit la colonne 'Estimated owners' en une valeur numérique moyenne."""
    if pd.isna(x):
        return np.nan
    
    # Nettoyer la chaîne
    x = str(x).strip().lower()
    
    # Gérer les formats incorrects
    if '-' not in x:
        return np.nan
    
    try:
        parts = x.split('-')
        low = int(parts[0].strip().replace(',', '').replace(' ', ''))
        high = int(parts[1].strip().replace(',', '').replace(' ', ''))
        return (low + high) / 2
    except:
        return np.nan
```

```python
# Création de la colnne numérique Estimated sales
df_analyse['Estimated sales'] = df_analyse['Estimated owners']
# Appliquer la conversion
df_analyse['Estimated sales'] = df_analyse['Estimated sales'].apply(parse_estimated_sales)
```
