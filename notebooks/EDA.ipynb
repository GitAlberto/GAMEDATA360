```python
# Fonction pour analyser les colonnes num√©riques
def analyse_colonne_numerique(df, col):
    """
    Analyse d'une colonne num√©rique :
    - Statistiques de base
    - Distribution (histogramme)
    - Boxplot pour les outliers
    """
    print(f"\n Analyse de la colonne : {col}")
    print("-" * 40)
    print(np.round(df[col].describe(),2))  # min, max, moyenne, m√©diane, etc.

    # D√©tection des valeurs aberrantes (IQR)
    Q1 = df[col].quantile(0.25) # le premier quartile
    Q3 = df[col].quantile(0.75) # le troisieme quartile
    IQR = Q3 - Q1 # difference interquartiles
    outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)] # dataframe contenant les enregisrements dont les "Price" sont ab√©rant
    print(f"\nNombre d'outliers d√©tect√©s : {len(outliers)}")

    # Pourcentage des valeurs manqantes
    nb_cells_col = len(df[col])
    nb_cells_nan = df[col].isna().sum()
    prctg_manq = (nb_cells_nan/nb_cells_col)*100
    print(f"Le pourcentage des valeurs manquantes est : {prctg_manq}%")
```

```python
# Analyse "Price"
analyse_colonne_numerique(data_copy,'Price')
```

```python
# Analyse de la colonne Peak CCU
analyse_colonne_numerique(data,"Peak CCU")
```

```python
# Analyse "Positive"
analyse_colonne_numerique(data_copy,"Positive")
```

```python
# Fonction d'analyse colonne objects
def analyse_colonne_categorielle(df,col):
    print(f"\n Analyse de la colonne : {col}")
    print("-" * 35) 
    print(df[col].describe(include = "object"))

    # Valeurs manquantes
    nb_cells_col = len(df[col])
    nb_cells_nan = df[col].isna().sum()
    prctg_manq = np.round((nb_cells_nan/nb_cells_col)*100,1)
    print(f"\nLe pourcentage des valeurs manquantes est : {prctg_manq}%")
    
```

```python
# Analys de la colonne genre
analyse_colonne_categorielle(data_copy,'Genres')
```

La plupart des jeux affich√©s pr√©sentent une anomalie dans leur nom, l'indication "Playtest" nous passons donc √† la suppression de celle-ci

```python
# Filtrage les lignes o√π 'Name' contient 'Playtest'
mask = data_copy["Name"].str.contains("Playtest", na=False)
df_playtest = data_copy[mask]
print("Nombre de lignes qui contiennent l'indication Playtest",len(df_playtest))
```

```python
# Supprission de l'indication 'Playtest' dans les noms filtr√©s
df_playtest["Name"] = df_playtest["Name"].str.replace("Playtest","").str.strip()
```

```python
# r√©injection dans le DataFrame original
data_copy.loc[mask, "Name"] = df_playtest["Name"]
```

```python
# V√©rification du n√©ttoyage des Playtest
data_copy[mask].head()
```

```python
# Standardisation des colonnes 'Windows', 'Mac', 'Linux' en 1 et 0
for col in ['Windows', 'Mac', 'Linux']:
    data_copy[col] = data_copy[col].map({True: 1, False: 0})
# V√©rification
data_copy[['Windows', 'Mac', 'Linux']].head()
```

```python
# Standardisation et retypage de la colonne 'Release date' en ann√©e/mois/jour
data_copy['Release date'] = pd.to_datetime(data_copy['Release date'], errors='coerce')
# V√©rification
data_copy['Release date'].head()
```

En constatant que les jeux sans recommendations sont aussi ceux sans developers et publishers,
on va cr√©er 2 nouveaux dataframes pour les jeux en version b√™ta et publi√©s, mais avant une colonne qui inquique si le jeu est publi√© ou pas s'il n'ont pas de publisers ni de developpers

```python
# Cr√©ation de la colonne 'published' : 0 si pas de Developers et Publishers, sinon 1
data_copy["published"] = data_copy.apply(
    lambda row: 0 if pd.isna(row["Developers"]) and pd.isna(row["Publishers"]) else 1,
    axis=1
)
data_copy["published"].head()
```
